```python
# Percentage of missing values
missing_percentages = (df.isnull().mean() * 100)
print(missing_percentages)  
```

```python
# Define the threshold (e.g., 30%)
threshold = 30

# Drop columns where the percentage of missing values exceeds the threshold
columns_to_drop = missing_percentages[missing_percentages > threshold].index

# Drop those columns from the dataframe
df = df.drop(columns=columns_to_drop)

print(f"Following columns were dropped: {list(columns_to_drop)}")
```

```python
# pd.factorize() can factorize categorical features to make them numeric (e.g. {"male": 0, "female": 1})
X_train["COUNTY_NAME_n"] = pd.factorize(X_train["COUNTY_NAME"])[0]
X_train["STATE_NAME_n"] = pd.factorize(X_train["STATE_NAME"])[0]
```

```python
# pd.factorize() can factorize categorical features to make them numeric (e.g. {"male": 0, "female": 1})
X_test["COUNTY_NAME_n"] = pd.factorize(X_test["COUNTY_NAME"])[0]
X_test["STATE_NAME_n"] = pd.factorize(X_test["STATE_NAME"])[0]
```

```python
# Drop original categorical columns
X_train.drop(["COUNTY_NAME", "STATE_NAME"], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit(X_train, X_test) 
```

```python
# Align the columns of X_test to match X_train
X_test = X_test[X_train.columns]
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)    
```

```python
from sklearn.feature_selection import f_classif, SelectKBest

selection_model = SelectKBest(f_classif, k = 5)
selection_model.fit(X_train_scaled, y_train)        
```

```python
ix = selection_model.get_support()

X_train_sel = pd.DataFrame(selection_model.transform(X_train_scaled), columns = X_train.columns.values[ix])
X_test_sel = pd.DataFrame(selection_model.transform(X_test_scaled), columns = X_test.columns.values[ix])    
```

```python
dict(zip(X_train.columns,selection_model.scores_))
```

```python
# Initialize the Lasso model with a chosen alpha (regularization strength)
lasso_model = Lasso(alpha=0.1)

# Train the model
lasso_model.fit(X_train, y_train)

# Predict on the test set
y_pred_lasso = lasso_model.predict(X_test)

# Evaluate the model
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("Lasso Regression - MSE:", mse_lasso)
print("Lasso Regression - R2:", r2_lasso)
```

```python
for alpha in [0.01, 0.1, 1, 10]:
    lasso_model = Lasso(alpha=alpha)
    lasso_model.fit(X_train, y_train)
    y_pred = lasso_model.predict(X_test)
    print(f"Alpha: {alpha} - MSE: {mean_squared_error(y_test, y_pred):.4f}, R2: {r2_score(y_test, y_pred):.4f}")    
```
